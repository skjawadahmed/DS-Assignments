```python
## Q1. Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.
# Create a list of 5 integers
numbers = [10, 20, 30, 40, 50]
print("Original list:", numbers)

# 1. append()
numbers.append(60)
print("After append(60):", numbers)

# 2. extend() 
numbers.extend([70, 80])
print("After extend([70, 80]):", numbers)

# 3. insert() 
numbers.insert(2, 25)
print("After insert(2, 25):", numbers)

# 4. remove() 
numbers.remove(40)
print("After remove(40):", numbers)

# 5. pop() 
popped_value = numbers.pop()   # removes the last item
print("After pop():", numbers)
print("Popped value:", popped_value)

# 6. clear() 
numbers_copy = numbers.copy()  # copy before clearing for next operations
numbers.clear()
print("After clear():", numbers)

# Restoring from copy for next operations
numbers = numbers_copy

# 7. index() 
index_of_25 = numbers.index(25)
print("Index of 25:", index_of_25)

# 8. count() 
count_of_20 = numbers.count(20)
print("Count of 20:", count_of_20)

# 9. sort() 
numbers.sort()
print("After sort():", numbers)

# 10. reverse() 
numbers.reverse()
print("After reverse():", numbers)
```

```python
# Q5. Write a script that checks if a person is eligible to vote (age ≥ 18). 
# Take age as a variable and print the appropriate message.

# Take age as a variable
age = int(input("Enter your age: "))

# Check voting eligibility
if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote yet.")

# Explanation:
# The input() function takes user input as a string, so we convert it to an integer using int().
# The if condition checks if the age is 18 or greater.
# If true, it prints that the person is eligible to vote.
# Otherwise, it prints that the person is not eligible yet.
```

```python
# Q6. Given a 'marks' variable, print the grade:
# - 90 and above: 'A'
# - 75–89: 'B'
# - 50–74: 'C'
# - Below 50: 'Fail'

# Take marks as a variable
marks = int(input("Enter your marks: "))

# Check the grade based on marks
if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: Fail")

# Explanation:
# The program checks the marks using if-elif-else conditions.
# - If marks are 90 or above, grade 'A' is printed.
# - If marks are between 75 and 89, grade 'B' is printed.
# - If marks are between 50 and 74, grade 'C' is printed.
# - If marks are below 50, 'Fail' is printed.
```

```python
# Q7. Given a number, check if it's positive, and if it is also even. 
# If not positive, print if it's zero or negative.

# Take a number as input
num = int(input("Enter a number: "))

# Check conditions
if num > 0:
    if num % 2 == 0:
        print("The number is positive and even.")
    else:
        print("The number is positive but odd.")
elif num == 0:
    print("The number is zero.")
else:
    print("The number is negative.")

# Explanation:
# The program first checks if the number is positive, zero, or negative.
# - If positive, it further checks whether it’s even (divisible by 2) or odd.
# - If the number equals zero, it prints that it’s zero.
# - Otherwise, it prints that the number is negative.
```

# Section C (Exploring Pandas)

```python
# Q11. Create a small DataFrame manually with Indian names and cities, then perform operations
import pandas as pd
import numpy as np

# Create DataFrame manually
data = {
    'Name': ['Amit', 'Priya', 'Rahul', 'Sneha', 'Rohit', 'Anjali', 'Vikram', 'Neha', 'Karan', 'Simran'],
    'Age': [25, 30, 22, 28, 35, 40, 27, 29, 31, 25],
    'City': ['Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Kolkata', 'Hyderabad', 'Pune', 'Jaipur', 'Lucknow', 'Ahmedabad'],
    'Salary': [50000, 60000, 55000, np.nan, 70000, 80000, 65000, 62000, 58000, np.nan]
}

df = pd.DataFrame(data)

# 1. Display information about the DataFrame
print("DataFrame info:")
print(df.info())
print("\nDataFrame description:")
print(df.describe())

# 2. Select 'Name' and 'City' columns
print("\nSelected columns ('Name' and 'City'):")
print(df[['Name', 'City']])

# 3. Drop 'City' column
df_dropped = df.drop('City', axis=1)
print("\nDataFrame after dropping 'City' column:")
print(df_dropped)

# 4. Fill missing values in 'Salary' with mean
salary_mean = df['Salary'].mean()
df['Salary'] = df['Salary'].fillna(salary_mean)
print("\nDataFrame after filling missing Salary values with mean:")
print(df)

# 5. Remove any duplicate rows
df_no_duplicates = df.drop_duplicates()
print("\nDataFrame after removing duplicates:")
print(df_no_duplicates)
```
