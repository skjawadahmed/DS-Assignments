```python
# 1. Load the patient dataset and show summary with info().
import pandas as pd

# Load datasets
patient_df = pd.read_csv("Patient_Data.csv")
billing_df = pd.read_csv("Billing_Data.csv")

# Dataset summary
patient_df.info()
```

```python
# 2. Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].
billing_relevant_df = patient_df[['PatientID', 'Department', 'Doctor', 'BillAmount']]
```

```python
# 3. Drop administrative columns like ['ReceptionistID', 'CheckInTime'].
patient_df = patient_df.drop(columns=['ReceptionistID', 'CheckInTime'])
```

```python
department_bill = patient_df.groupby('Department')['BillAmount'].sum()
print(department_bill)
```

```python
patient_df = patient_df.drop_duplicates(subset='PatientID')
```

```python
mean_bill = patient_df['BillAmount'].mean()
patient_df['BillAmount'] = patient_df['BillAmount'].fillna(mean_bill)
```

```python
merged_df = pd.merge(patient_df, billing_df, on='PatientID', how='inner')
```

```python
new_patients_df = pd.DataFrame({
    'PatientID': [201, 202],
    'Name': ['Ayaan', 'Meera'],
    'Department': ['Cardiology', 'Neurology'],
    'Doctor': ['Dr. Khan', 'Dr. Sharma'],
    'BillAmount': [15000, 18000],
    'InsuranceCovered': ['Yes', 'No'],
    'FinalAmount': [12000, 18000]
})

final_df = pd.concat([merged_df, new_patients_df], axis=0, ignore_index=True)
```

```python
extra_billing_columns = pd.DataFrame({
    'DiscountApplied': [500, 700],
    'PaymentMode': ['Cash', 'Card']
})

final_df = pd.concat([final_df, extra_billing_columns], axis=1)
```
