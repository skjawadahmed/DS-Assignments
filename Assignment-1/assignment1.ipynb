```python
#EXCERISE 1: Prime Numbers  
#Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.
def is_prime(n):
    # Prime numbers are greater than 1
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:  
            return False
    return True

# Take input from user
num = int(input("Enter a number: "))

# Check and display result
if is_prime(num):
    print(f"{num} is a Prime Number.")
else:
    print(f"{num} is NOT a Prime Number.")
```

```python
#Exercise 2: Product of Random Numbers
#Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.

import random

num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

print(f"Multiply: {num1} × {num2}")

user_answer = int(input("Enter your answer: "))
# Checking
if user_answer == num1 * num2:
    print("Correct Answer!.")
else:
    print(f"Incorrect. The correct answer is {num1 * num2}.")
```

```python
# Exercise 3: Squares of Even Numbers from 100 to 200

# Choice: EVEN numbers
print("Squares of EVEN numbers between 100 and 200:\n")

for num in range(100, 201): 
    if num % 2 == 0:  # Check if number is even
        print(f"{num}² = {num ** 2}")
```

```python
# Exercise 4: Hospital Billing System

# Fixed charges
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_charge = 300

# Patient details
days_stayed = 4
num_lab_tests = 3
medicine_charges = 2400

# Calculations
room_total = room_charge_per_day * days_stayed
lab_total = lab_test_charge * num_lab_tests

# Total before discount
total_before_discount = room_total + doctor_fee + lab_total + medicine_charges

# Apply discount if applicable
if total_before_discount > 10000:
    discount = total_before_discount * 0.10
else:
    discount = 0

final_bill = total_before_discount - discount

# Output
print("------ Hospital Billing System ------")
print(f"Room Charges (₹{room_charge_per_day} × {days_stayed} days): ₹{room_total}")
print(f"Doctor Consultation Fee: ₹{doctor_fee}")
print(f"Lab Test Charges (₹{lab_test_charge} × {num_lab_tests} tests): ₹{lab_total}")
print(f"Medicine Charges: ₹{medicine_charges}")
print(f"Total before discount: ₹{total_before_discount}")
print(f"Discount Applied: ₹{discount}")
print(f"Final Bill Amount: ₹{final_bill}")
```
