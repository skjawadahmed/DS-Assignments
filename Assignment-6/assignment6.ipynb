# a) Estimation and confidence intervals

```python
# necessary libraries
import numpy as np
from scipy import stats
```

```python
# Step 2: Load the Given Data

# Durability data of print-heads (in millions of characters)
data = np.array([
    1.13, 1.55, 1.43, 0.92, 1.25,
    1.36, 1.32, 0.85, 1.07, 1.48,
    1.20, 1.33, 1.18, 1.22, 1.29
])
```

```python
# Step 3: Calculate Sample Statistics
# Number of observations in the sample
n = len(data)

# Calculate the sample mean
mean = np.mean(data)

# Calculate the sample standard deviation
# ddof=1 is used because this is a sample (not population)
std_sample = np.std(data, ddof=1)

# Display computed statistics
print("Sample size (n):", n)
print("Sample mean:", mean)
print("Sample standard deviation:", std_sample)
```

```python
# PART (a): 99% Confidence Interval (σ Unknown → t-Distribution)

# Set the confidence level to 99%
confidence_level = 0.99

# Alpha value (significance level)
alpha = 1 - confidence_level

# Degrees of freedom (n - 1) for t-distribution
df = n - 1

# Compute t critical value for 99% confidence interval
t_critical = stats.t.ppf(1 - alpha / 2, df)

# Calculate standard error using sample standard deviation
standard_error_t = std_sample / np.sqrt(n)

# Compute lower and upper limits of the confidence interval
ci_lower_t = mean - t_critical * standard_error_t
ci_upper_t = mean + t_critical * standard_error_t

# Display results
print("\n99% Confidence Interval using t-distribution:")
print("Lower Limit:", ci_lower_t)
print("Upper Limit:", ci_upper_t)
```

```python
# PART (b): 99% Confidence Interval (σ Known → z-Distribution)

# Given population standard deviation
sigma = 0.2

# Compute z critical value for 99% confidence interval
z_critical = stats.norm.ppf(1 - alpha / 2)

# Calculate standard error using population standard deviation
standard_error_z = sigma / np.sqrt(n)

# Compute lower and upper limits of the confidence interval
ci_lower_z = mean - z_critical * standard_error_z
ci_upper_z = mean + z_critical * standard_error_z

# Display results
print("\n99% Confidence Interval using z-distribution:")
print("Lower Limit:", ci_lower_z)
print("Upper Limit:", ci_upper_z)
```

# b) HYPOTHESIS TESTING

```python

import math
# Given data
sample_mean = 3050      # Sample mean weekly cost (Rs.)
n = 25                  # Sample size
alpha = 0.05            # Significance level

# Distribution of units produced (X)
mean_X = 600
sd_X = 25

# Hypotheses
# H0: μ = 1000 + 5X  (cost follows the model)
# H1: μ > 1000 + 5X (cost is higher than the model)

# Theoretical mean and standard deviation of weekly cost
mu = 1000 + 5 * mean_X
sigma = 5 * sd_X

# Standard error
standard_error = sigma / math.sqrt(n)

# Z-test statistic
z_statistic = (sample_mean - mu) / standard_error

# Critical value for right-tailed test at α = 0.05
critical_z = 1.645

# Standard normal cumulative distribution function
def normal_cdf(z):
    return 0.5 * (1 + math.erf(z / math.sqrt(2)))

# P-value (right-tailed)
p_value = 1 - normal_cdf(z_statistic)

# Decision rule
if z_statistic > critical_z:
    decision = "Reject Null Hypothesis"
else:
    decision = "Fail to Reject Null Hypothesis"

# Conclusion
if decision == "Reject Null Hypothesis":
    conclusion = "There is strong evidence that weekly operating costs are higher than the theoretical model."
else:
    conclusion = "There is no sufficient evidence to support the claim that weekly operating costs are higher than the model."

# Output
print("Hypothesis Testing Results")
print("Theoretical Mean (μ):", mu)
print("Standard Deviation (σ):", sigma)
print("Sample Mean (x̄):", sample_mean)
print("Sample Size (n):", n)
print("Z-statistic:", round(z_statistic, 2))
print("Critical Z-value:", critical_z)
print("P-value:", round(p_value, 6))
print("Decision:", decision)
print("Conclusion:", conclusion)
```
